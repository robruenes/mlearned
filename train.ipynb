```python
import os
import pandas as pd
```

```python
categories_transformation = {"CURR EVENTS": "0",
                             "WORLD HIST": "1",
                             "AMER HIST": "2",
                             "LIFESTYLE": "3",
                             "BUS/ECON": "4",
                             "CLASS MUSIC": "5",
                             "ART": "6",
                             "FILM": "7",
                             "THEATRE": "8",
                             "LANGUAGE": "9",
                             "TELEVISION": "10",
                             "LITERATURE": "11",
                             "FOOD/DRINK": "12",
                             "GEOGRAPHY": "13",
                             "POP MUSIC": "14",
                             "GAMES/SPORT": "15",
                             "SCIENCE": "16",
                             "MATH": "17"
                             }

outcome_transformation = {"W": "2", "T": "1", "L": "0"}
rundle_transformation = {"A": "5", "B": "4", "C": "3", "D": "2", "E": "1", "R": "0"}
```

```python
player_dfs = {}
season_dfs = {}

for path, dirs, files in os.walk("data"):
    if "player" not in path:
        continue

    id = path.split("player_")[1]
    if id not in player_dfs:
        player_dfs[id] = {}

    for f in files:
        if f == "categorical_stats.csv":
            categorical_df = pd.read_csv(os.path.join(path, f))
            categorical_df = categorical_df.replace(
                to_replace=categories_transformation
            )
            player_dfs[id]["categorical"] = categorical_df
            continue

        if ".csv" not in f:
            continue

        season = f.split(".csv")[0]
        matches_df = pd.read_csv(os.path.join(path, f))
        matches_df = matches_df.replace(to_replace=rundle_transformation)
        matches_df = matches_df.replace(to_replace=outcome_transformation)
        player_dfs[id][season] = matches_df
        # if "matches" not in player_dfs[id]:
        #     player_dfs[id]["matches"] = matches_df
        #     continue

        # player_dfs[id]["matches"] = pd.concat([player_dfs[id]["matches"], matches_df], axis=0)
        # player_dfs[id]["matches"].reset_index(inplace=True, drop=True)
```

```python
season_questions = {}
for path, _, files in os.walk("data/seasons"):
    for f in files:
        season_df = pd.read_csv(os.path.join(path, f))
        # Need to do this because the match category files currently include whitespace.
        season_df = season_df.map(lambda x: categories_transformation[x.strip()])
        season = f.split("match_categories_")[1].split(".csv")[0]
        season_questions[season] = season_df

print(season_questions)
```

```python
# only need to pull categorical stats for them.
unique_opponents = set()
for player in player_dfs:
  for key, df in player_dfs[player].items():
    if key == "categorical":
      continue
    
    # print(list(df[["Opponent"]]))
    unique_opponents.update(df["Opponent"].to_list())
```

```python
import json

opponent_dict = {}

for opponent in unique_opponents:
    opponent_dict[opponent] = {"name": f"{opponent}"}

with open("opponents.json", "w") as ofile:
    json.dump(opponent_dict, ofile)
```
